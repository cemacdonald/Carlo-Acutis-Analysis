```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import arrow
```

```python
# Extract events over time for each country
data = pd.read_csv('eucharistic-miracles-country.csv')
countries = ['Argentina','Austria', 'Belgium', 'Colombia', 'Croatia', 'Egypt', 'France',
             'Germany', 'Martinique', 'Reunion Islands', 'Italy', 'Mexico','Holland', 
             'Peru', 'Poland', 'Portugal', 'Spain', 'Switzerland','Venezuela']

colors = ['#ff5733', '#ffc300', '#daf7a6', '#ffa07a', '#9a7d0a', '#00bfff', '#ff00ff', '#7fffd4',
          '#40e0d0', '#0000ff', '#ff1493', '#008000', '#800080', '#ffff00', '#ffd700', '#ffa500',
          '#c0c0c0', '#800000', '#000000']

timeframe = {}
for country in countries:
    years = data[country].dropna().astype(int)
    timeframe[country] = [arrow.get(year, 1, 1) for year in years]

# Plot a histogram for each country
plt.figure(figsize=(18, 9))
index= 0
for country, events in timeframe.items():
    plt.hist([event.year for event in events], bins=range(1500,2050, 3),alpha = .75 , label=country, color=colors[index])
    plt.yticks(np.arange(0, plt.gca().get_ylim()[1]+1, 1))
    index += 1

# Customize the plot
plt.title('Frequency of Eurcharistic Miracles from the years 1500-2024')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
